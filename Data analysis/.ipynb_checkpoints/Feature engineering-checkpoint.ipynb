```python
for i in tqdm(range(len(position))):
    SUM1=0
    SUM2=0
    for j in range(len(top_tier)):
        if position.loc[:,'top'][i]==top_tier.loc[:,'champid'][j]:
            SUM1+=top_tier.loc[:,'tier'][j]
        if position.loc[:,'count_top'][i]==top_tier.loc[:,'champid'][j]:
            SUM2+=top_tier.loc[:,'tier'][j]  
            
    for j in range(len(jungle_tier)):
        if position.loc[:,'jungle'][i]==jungle_tier.loc[:,'champid'][j]:
            SUM1+=top_tier.loc[:,'tier'][j]
        if position.loc[:,'count_jungle'][i]==jungle_tier.loc[:,'champid'][j]:
            SUM2+=top_tier.loc[:,'tier'][j]
            
    for j in range(len(mid_tier)):
        if position.loc[:,'mid'][i]==mid_tier.loc[:,'champid'][j]:
            SUM1+=mid_tier.loc[:,'tier'][j]
        if position.loc[:,'count_mid'][i]==mid_tier.loc[:,'champid'][j]:
            SUM2+=mid_tier.loc[:,'tier'][j]
            
    for j in range(len(carry_tier)):
        if position.loc[:,'carry'][i]==carry_tier.loc[:,'champid'][j]:
            SUM1+=carry_tier.loc[:,'tier'][j]
        if position.loc[:,'count_carry'][i]==carry_tier.loc[:,'champid'][j]:
            SUM2+=carry_tier.loc[:,'tier'][j]
            
    for j in range(len(support_tier)):
        if position.loc[:,'support'][i]==support_tier.loc[:,'champid'][j]:
            SUM1+=support_tier.loc[:,'tier'][j]
        if position.loc[:,'count_support'][i]==support_tier.loc[:,'champid'][j]:
            SUM2+=support_tier.loc[:,'tier'][j]
    position['B_tier'][i]=SUM1/5
    position['R_tier'][i]=SUM2/5
```
