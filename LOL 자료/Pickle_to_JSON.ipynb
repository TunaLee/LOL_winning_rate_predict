```python
match_df.drop(columns = ['gameCreation', 'gameDuration', 'queueId', 'gameVersion'], axis=1, inplace=True)
```

```python
match_df_team100 = []
match_df_team200 = []
match_df_team100_participant = []
match_df_team200_participant = []
for i in range(len(match_df)):
    match_df_team100.append((match_df['teams'][i][0]['teamId'], match_df['teams'][i][0]['win'] ))
    match_df_team200.append((match_df['teams'][i][1]['teamId'], match_df['teams'][i][1]['win'] ))
    team100_participant = []
    team200_participant = []
    for j in range(5):
        team100_participant.append((match_df['participantIdentities'][i][j]['player']['summonerName'], match_df['participants'][i][j]['championId']))
        team200_participant.append((match_df['participantIdentities'][i][j+5]['player']['summonerName'], match_df['participants'][i][j+5]['championId']))
    match_df_team100_participant.append(team100_participant)
    match_df_team200_participant.append(team200_participant)
```

```python
match_df_team100 = pd.Series(match_df_team100, index = match_df.index)
match_df_team200 = pd.Series(match_df_team200, index = match_df.index)
match_df_team100_participant = pd.Series(match_df_team100_participant, index = match_df.index)
match_df_team200_participant = pd.Series(match_df_team200_participant, index = match_df.index)
match_df0 = pd.concat((match_df, match_df_team100, match_df_team200, match_df_team100_participant, match_df_team200_participant), axis=1)
```

```python
match_df0.columns = [*match_df.columns, 'teamBlue', 'teamRed', 'teamBlueParticipants', 'teamRedParticipants'] 
```

```python
match_df0.drop(columns=['teams', 'participants', 'participantIdentities'], axis=1, inplace=True)
```

```python
match_df0.to_json('Gameinfo_70000_to_110000_parsed.json')
```
