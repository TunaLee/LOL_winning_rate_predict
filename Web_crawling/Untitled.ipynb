```python
champ_list = pd.read_csv('Champ.csv').iloc[:,1:]
```

```python
champ_list = champ_list.sort_values('number')
```

```python
champ_list['name']=''
```

```python
champ_list_kor.keys()
```

```python
for i in tqdm(range(len(champ_list))):
    for j in range(len(champ_list_kor)):
        if champ_list.loc[:,'number'][i]==int(list(champ_list_kor.keys())[j]):
            champ_list.loc[:,'name'][i]=list(champ_list_kor.values())[j]
```
